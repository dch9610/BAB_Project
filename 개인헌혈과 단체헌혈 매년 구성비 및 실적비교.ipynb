```python
blood_place2 = blood_place.T.drop(columns = [4,5,6,58,59,60])
blood_place3 = blood_place2[blood_place2[2].str.contains('소계|성별')]
blood_place4 = blood_place3[blood_place3[1].str.contains('헌혈|성별')]
blood_place4
```

### 단체헌혈 실적 확인

```python
blood_group  = blood_place4[blood_place4[1].str.contains('단체')]
blood_group2 = blood_group[blood_group[3].str.contains('실적')][3:]
blood_group2
```

```python
blood_group3 = blood_group2[[0,7]]
blood_group3 = blood_group3.astype(int)
blood_group3 =blood_group3.set_index(0)
blood_group3.columns= ['단체 헌혈 실적']
```

```python
blood_group3.plot()
plt.title('년도별 단체헌혈 실적 변화')
plt.show()
```

## 개인 헌혈실적 확인

```python
blood_one  = blood_place4[blood_place4[1].str.contains('개인')]
blood_one2 = blood_one[blood_one[3].str.contains('실적')][3:]
blood_one2
```

```python
blood_one3 = blood_one2[[0,7]]
blood_one3 = blood_one3.astype(int)
blood_one3 =blood_one3.set_index(0)
blood_one3.columns= ['개인 헌혈 실적']
blood_one3
```

```python
blood_one3.plot()
plt.title('년도별 개인헌혈 실적 변화')
plt.show()
```

## 개인이랑 단체를 실적을 합쳐서 보자

```python
blood_result =pd.concat([blood_one3,blood_group3], axis=1)
blood_result
```

```python
blood_result.plot()
plt.title('개인, 단체 실적비교')
plt.show()
```

개인과 단체 실적을 비교해보면 단체 실적은 크게 변화가 없었음<br>
전체 헌혈 실적의 증가는 개인의 헌햘횟수나 양이 증가한 것으로 파악된다.

```python
blood_result['헌혈 실적'] = blood_result['개인 헌혈 실적'] + blood_result['단체 헌혈 실적']
blood_result
```
