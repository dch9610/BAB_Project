## 결과값 데이터

```python
# 결과값으로 쓸 데이터를 가져온다
blood = pd.read_csv('blood data/수혈용_혈액_월평균_보유_추이_2005_2019.csv', encoding='CP949')
blood
```

```python
# 데이터형태를 바꿔서 필요한 데이터를 뽑기 쉽게한다.
blood_co=blood.T
blood_co=blood_co.rename(columns=blood_co.iloc[0])
blood_co
```

```python
# 농축적혈구가 결과값 기준이니 농축적혈구가 월별 농축적혈구 값만 가져온다.
blood_co2 = blood_co[blood_co['월별(1)'].str.contains('농축적혈구')].drop(columns='합계')
```

```python
# 2008년부터의 데이터를 쓸거니 불필요한 데이터를 제거해준다
blood_co2.drop(index=['2005','2006','2007'], inplace=True)
```

```python
# 데이터를 한번 저장해준다
blood_co2.to_csv('전처리 데이터/적혈구 기준 월별 보유 추이.csv' ,encoding='CP949')
```

```python
blood_co2
```

```python
# 불필요한 값 제거
blood_co3= blood_co2.T.drop(index=('월별(1)'))
blood_co3
```

```python
# 데이터 프레임으로 만들기위해 그냥 print 찍어버림
# 각 년도별 데이터
for i in blood_co3.columns:
    print(f'blood_co3["{i}"]')
```

```python
# 각년도별 데이터를 가져와서 합쳐준다
a1 = pd.concat([blood_co3["2008"],
blood_co3["2009"],
blood_co3["2010"],
blood_co3["2011"],
blood_co3["2012"],
blood_co3["2013"],
blood_co3["2014"],
blood_co3["2015"],
blood_co3["2016"],
blood_co3["2017"],
blood_co3["2018"],
blood_co3["2019"]])
```

```python
# 합치기위해 데이터 프레임화시킨다.
a2 =pd.DataFrame({'volume' : a1})
a2
```

```python
# 데이터를 따로 저장해준다
a2.to_csv('모델용 데이터/월별 적혈구 기준 보유량(model).csv', encoding='CP949')
```

```python
# 합치기 편하게 하기위해 공통컬럼을 만들어준다
re_base_co = base_co.reset_index()
a2 = a2.reset_index().drop(columns= 'index').reset_index()

display(re_base_co)
display(a2)
```

```python
# 데이터 프레임 두개를 합친다.
df = pd.merge(re_base_co, a2, on='index')
df
```

```python
# 합친 데이터프레임을 저장해준다
df.to_csv('모델용 데이터/모델링데이터(1).csv', encoding= 'CP949')
```

## 혈액 공급 실적

```python
# 공급량을 가져오기위해 데이터 가져오기
blood_sup = pd.read_csv('blood data/혈액공급실적_2005_2019.csv', encoding='CP949')
blood_sup
```

```python
# 데이터를 편하게 변환
blood_sup_co= blood_sup.T
blood_sup_co2 = blood_sup_co.rename(columns=blood_sup_co.loc['혈액원별(1)'] )
blood_sup_co2
```

```python
# 년도별 공급합계 데이터만 가져오기
blood_sup_co3 = blood_sup_co2[blood_sup_co[0].str.contains('공급합계')]
blood_sup_co3
```

```python
# 공급합계중 지역 합계 데이터만 쓸것이다
blood_sup_co4 = blood_sup_co3[['합계']]
# 2008년 이후 데이터만  쓸거니 필요없는건 drop
blood_sup_co5 = blood_sup_co4.drop(index=['2005.2','2006.2','2007.2'])
blood_sup_co5
```

```python
# 합치는거 명령어 적기 편하게 12개 프린트로 만들어서 복사
for i in range(12):
    print(i)
    print(f'blood_sup_co5')
```

```python
# 년단위 데이터를 월 단위에 넣어주기위해 데이터 처리
# 데이터를 12번 반복하기위해 같은데이터를 12번 합침
blood_sup_co6 = pd.concat([blood_sup_co5,
blood_sup_co5,
blood_sup_co5,
blood_sup_co5,
blood_sup_co5,
blood_sup_co5,
blood_sup_co5,
blood_sup_co5,
blood_sup_co5,
blood_sup_co5,
blood_sup_co5,
blood_sup_co5])
blood_sup_co6
```

```python
# 순서대로 넣기위해 정렬
blood_sup_co6 = blood_sup_co6.sort_index()
blood_sup_co6 = blood_sup_co6.rename(columns={'합계' : 'sup_result'})
blood_sup_co6
```

```python
# 데이터 저장해주기
blood_sup_co6.to_csv('전처리 데이터/혈액공급합계.csv', encoding='CP949')
```

```python
# 합치기위해 index 칼럼 넣어주기
blood_sup_co7 = blood_sup_co6.reset_index().drop(columns= 'index')
blood_sup_co7 = blood_sup_co7.reset_index()
blood_sup_co7
```
