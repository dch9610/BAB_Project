```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import data
import datetime
import seaborn as sns
# 그래프 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
# plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['font.size'] = 16
plt.rcParams['figure.figsize'] = 20, 10
plt.rcParams['axes.unicode_minus'] = False
```

```python
blood = pd.read_csv('blood data/수혈용_혈액_보유_추이_2005_2019.csv', encoding = 'cp949')
blood
```

```python
blood2= blood.set_index('일별(1)')
blood2
```

```python
blood2[blood2['2006'] == '-'].index
```

```python
blood2 = blood2.drop(blood2[blood2['2006'] == '-'].index)
blood2
```

```python
blood2 = blood2.astype(int)
```

```python
blood2.info()
```

```python
for i in blood2.columns:
    plt.plot(blood2.index, blood2[f'{i}'])
    plt.title(f'{i}년도 헌혈 보유 추이')
    plt.show()
```

```python
blood2.sum()
```

```python
blood2.sum().max()
```

```python

plt.bar(blood2.columns , blood2.mean(), color = sns.color_palette('hls',len(blood2.columns)))
plt.title('년도별 혈액 보유량')
plt.ylim(0, blood2.mean().max())
plt.xlabel('년도')
plt.ylabel('혈액 보유 추이')
plt.show()
# plt.savefig('년도별 혈약 보유량 알록달록')
```

```python

plt.bar(blood2.columns , blood2.sum(), color = 'lightblue')
plt.title('년도별 혈액 보유량')
plt.ylim(0, blood2.sum().max()+1000000)
plt.xlabel('년도')
plt.ylabel('혈액 보유 추이')
plt.show()
```
