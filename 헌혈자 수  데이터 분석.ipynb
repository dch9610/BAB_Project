```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import data
import datetime
import seaborn as sns
# 그래프 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
# plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['font.size'] = 16
plt.rcParams['figure.figsize'] = 20, 10
plt.rcParams['axes.unicode_minus'] = False
```

```python
blood = pd.read_csv('blood data/헌혈자_수_2005_2019.csv', encoding = 'cp949')
blood
```

```python
blood_map = blood[['헌혈자 수 (명)']]
blood_per = blood[['헌혈률 (%)','현혈가능인구대비 헌혈률 (%)']]
```

```python
# 년도별 헌혈자 수 데이터
plt.plot( blood['시점'],blood_map, 'bo--')

plt.rcParams['font.size'] = 20
plt.title('년도별 헌혈자 수')
plt.show()
plt.savefig('년도별 헌혈자 수.png')
```

```python
# 년도별 헌혈률 그래프
plt.plot( blood['시점'],blood['헌혈률 (%)'], 'b', label='헌혈률')
plt.plot( blood['시점'],blood['현혈가능인구대비 헌혈률 (%)'], 'r', label='헌혈가능인구 헌혈률')

plt.title('헌혈률그래프')
plt.legend()
plt.show()
```
